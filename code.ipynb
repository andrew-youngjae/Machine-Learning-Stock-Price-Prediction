```python
# training
# code block 1 : 데이터로부터 각종 금융 지표 추출하기
from tqdm.notebook import tqdm
import scipy
import numpy as np
import pandas as pd
import os

def weighted_mean(weight_array):
    def inner(x):
        return (weight_array * x).mean()
    return inner

# 시장 데이터 가져오기
df_feature_market = pd.read_csv(os.path.join('data', 'training', 'training_market', 'feature', 'feature_market.csv'), index_col=False, converters={'date': lambda x: str(x)})

# 시장 데이터 전처리 - nan을 제거하거나 적절하게 대체
df_training_market = df_feature_market.fillna(method='ffill')
df_training_market = df_training_market.fillna(0)

print(df_training_market.shape)

# trainingdata 폴더 만들고 저장
file_path = os.path.join('data', 'training', 'training_market', 'trainingdata', 'trainingdata_market.csv')
os.makedirs(os.path.dirname(file_path), exist_ok=True)
df_training_market.to_csv(file_path, index=False)

# 주식 feature 데이터 가져오기
for path, dirs, files in os.walk(os.path.join('data', 'training', 'training_stock', 'feature')):
    pbar = tqdm(files)
    for filename in pbar:
        code = filename.split('.')[0].split('_')[-2]
        name = filename.split('.')[0].split('_')[-1]
        pbar.set_description(code)
        file_path = os.path.join(path, filename)
        df_feature = pd.read_csv(file_path, index_col=False, converters={k: lambda x: str(x) for k in ['code', 'name', 'date']})

        # preporcess
        df_training = df_feature.fillna(method='ffill')
        df_training = df_training.fillna(0)

        # find peaks - 레이블 생성을 위해 종가 차트에서 peak를 찾아내기
        peaks = []
        peaks.extend(scipy.signal.find_peaks(df_training['close'], distance=5, width=10)[0])
        peaks.extend(scipy.signal.find_peaks(-df_training['close'], distance=5, width=10)[0])
        if len(df_training)-1 not in peaks:
            peaks.append(len(df_training)-1)

        # 수익성, 안전성, 유동성 지표 추가 - peak_date, peak_close는 수익성/위험성/유동성 지표 구하기 위해 임시로 생성
        df_training.loc[:, 'peak_date'] = ''
        df_training.loc[:, 'peak_close'] = np.nan
        df_training.loc[:, 'peak_diffratio'] = np.nan
        df_training.loc[:, 'interpeak_mdd'] = np.nan
        df_training.loc[:, 'interpeak_trans_price_exp'] = np.nan
        
        # ema start
        df_training['ema5'] = df_training['close'].ewm(5).mean()
        df_training['ema20'] = df_training['close'].ewm(20).mean()
        df_training['ema100'] = df_training['close'].ewm(100).mean()
        df_training['ema200'] = df_training['close'].ewm(200).mean()
        # ema end
        
        # wma start
        weights = np.arange(1,6)
        wma5 = df_training['close'].rolling(5).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
        
        weights = np.arange(1,21)
        wma20 = df_training['close'].rolling(20).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
        
        weights = np.arange(1,101)
        wma100 = df_training['close'].rolling(100).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)
        
        weights = np.arange(1,201)
        wma200 = df_training['close'].rolling(200).apply(lambda prices: np.dot(prices, weights)/weights.sum(), raw=True)

        df_training['wma5'] = wma5
        df_training['wma20'] = wma20
        df_training['wma100'] = wma100
        df_training['wma200'] = wma200
        # wma end
        
        # macd start
        # 단기(window = 12) 지수 이동 평균선
        short_ema = df_training['close'].ewm(span = 12, adjust = False).mean()

        # 장기(window = 26) 지수 이동 평균선
        long_ema = df_training['close'].ewm(span = 26, adjust = False).mean()

        macd_line = short_ema - long_ema

        # macd column 추가
        df_training['macd'] = macd_line

        #signal(window = 9) column 추가
        df_training['signal'] = df_training['macd'].ewm(span = 9, adjust = False).mean()

        # macd 선이 signal 선을 상향 돌파 -> 매수 사인
        # macd 선이 signal 선을 하향 돌파 -> 매도 사인

        # 빈 데이터 채워넣기
        #df_training = df_training.fillna(method='bfill')
        # macd end
        
        # for bb
        df_training.loc[:, 'bb_middle'] = np.nan
        df_training.loc[:, 'stddev'] = np.nan
        df_training.loc[:, 'bb_upper'] = np.nan
        df_training.loc[:, 'bb_lower'] = np.nan
        df_training.loc[:, 'bb_bandwidth'] = np.nan
        
        # bb start        
        df_training.loc[:, 'bb_middle'] = df_training.loc[:, 'close'].rolling(window=20).mean()
        df_training.loc[:, 'stddev'] = df_training.loc[:, 'close'].rolling(window=20).std()
        df_training.loc[:, 'bb_upper'] = df_training.loc[:, 'bb_middle'] + (df_training.loc[:, 'stddev'] * 2)
        df_training.loc[:, 'bb_lower'] = df_training.loc[:, 'bb_middle'] - (df_training.loc[:, 'stddev'] * 2)
        df_training.loc[:, 'bb_bandwidth'] = (df_training.loc[:, 'close'] - df_training.loc[:, 'bb_lower']) / df_training.loc[:, 'bb_middle'] * 100

        df_training = df_training.fillna(method='bfill')
        
        df_training = df_training.drop(['stddev'], axis=1)        
        # bb end

        #daily_return start
        df_training['daily_return']=df_training['close'].pct_change()*100
        #daily_return end
        
        #수익률 표준편차 start
        rolling_window=20
        df_training['stddev_returns'] = df_training['daily_return'].rolling(window=rolling_window).std()
        #수익률 표준편차 end

        _last_date = ''
        for peak in peaks:
            _date = df_training.iloc[peak]['date']
            _close = df_training.iloc[peak]['close']
            mask = (df_training['date'] >= _last_date) & (df_training['date'] <= _date)
            _last_date = _date
            if len(df_training[mask]) > 0:
                df_training.loc[mask, 'peak_date'] = _date
                df_training.loc[mask, 'peak_close'] = _close
                _x = np.array(df_training.loc[mask, 'close'])
                lower = np.argmax(np.maximum.accumulate(_x) - _x)
                upper = np.argmax(_x[:lower+1])
                
                # 위험성 지표 - 다음 피크까지의 최대 낙폭(MDD)
                df_training.loc[mask, 'interpeak_mdd'] = (_x[lower] - _x[upper]) / _x[upper]
                
                # 유동성 지표 - 다음 피크까지의 거래대금 예측치의 평균
                df_training.loc[mask, 'interpeak_trans_price_exp'] = df_training.loc[mask, 'trans_price_exp'].mean()

        # 수익성 지표 - 현재의 종가와 다음 피크와의 종가 비율
        df_training.loc[:, 'peak_diffratio'] = (df_training.loc[:, 'peak_close'] - df_training.loc[:, 'close']) / df_training.loc[:, 'close']

        # peak_date, peak_close feature는 이제 필요없으니 제거
        df_training = df_training.drop(['peak_date', 'peak_close'], axis=1)

        # trainingdata 파일에 저장
        file_path = os.path.join('data', 'training', 'training_stock', 'trainingdata', f'trainingdata_{code}_{name}.csv')
        os.makedirs(os.path.dirname(file_path), exist_ok=True)
        df_training.to_csv(file_path, index=False)
```

```python
# code block 5 : construct model
import torch
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = torch.nn.Sequential(
    torch.nn.Linear(input_dim, 256),
    torch.nn.ReLU(),
    torch.nn.Dropout(p=0.1),
    torch.nn.Linear(256, 128),
    torch.nn.ReLU(),
    torch.nn.Dropout(p=0.1),
    torch.nn.Linear(128, 64),
    torch.nn.ReLU(),
    torch.nn.Dropout(p=0.1),
    torch.nn.Linear(64, 32),
    torch.nn.ReLU(),
    torch.nn.Dropout(p=0.1),
    torch.nn.Linear(32, 1),
    torch.nn.Sigmoid(),
)
def init_weights(m):
    if isinstance(m, torch.nn.Linear):
        torch.nn.init.normal_(m.weight, std=0.01)
model.apply(init_weights)
model.to(device)
```

```python
# code block 6 : training model

print(label_type)
for epoch in tqdm(range(10)):
    for path, dirs, files in os.walk(os.path.join('data', 'training', 'training_stock', 'X')):
        loss = 0
        for filename in tqdm(files):
            code = filename.split('.')[0].split('_')[-2]
            name = filename.split('.')[0].split('_')[-1]
            # feature 데이터 X 가져오기
            file_path = os.path.join(path, filename)
            X = np.load(file_path)
            # label y 가져오기
            file_path = os.path.join(os.path.dirname(path), 'y', f'y_{code}_{name}.npy')
            y = np.load(file_path)
            
            X = torch.from_numpy(X).float().to(device)
            y = torch.from_numpy(y).float().to(device)
            
            criterion = torch.nn.MSELoss()
            optimizer = torch.optim.RMSprop(model.parameters(), lr=0.0001)
            optimizer.zero_grad()

            # 모델에 feature X 넣어 y_pred(예측값) 도출
            y_pred = model(X)

            # y_pred(예측값)와 y(정답값)의 차이(loss)를 줄여가는 방향으로 모델을 학습시키기
            _loss = criterion(y_pred[:, 0], y)
            _loss.backward()
            optimizer.step()
            _loss = _loss.item()
            loss += _loss
        print(f'epoch={epoch}\tloss={loss}')

# 학습이 완료된 모델을 파일로 저장
os.makedirs('models', exist_ok=True)
path = os.path.join('models', 'model.mdl')
torch.save(model, path)
```

```python
import os
import time
from pytz import timezone
from datetime import datetime, timedelta
import json
import argparse

import numpy as np
import pandas as pd
from tqdm import tqdm


FORMAT_DATE = '%Y%m%d'
start_date = '20220401'
end_date = '20220527'
dt_start_date = datetime.strptime(start_date, FORMAT_DATE)
dt_end_date = datetime.strptime(end_date, FORMAT_DATE)

MAX_HISTORY = 5


def convert_str2date(date):
    return datetime.strptime(date, FORMAT_DATE)


def get_today():
    dt = datetime.datetime.fromtimestamp(time.time(), timezone('Asia/Seoul'))
    date = dt.date()
    return date


class Backtest:
    def __init__(self, start_date=None, end_date=None, 
        initial_balance=100000000, max_investment=10000000, max_holdings=10, max_holdings_value=100000000, 
        rank_threshold_buy=5, rank_threshold_hold=1000, stoploss=-0.3, stopprofit=1.0, holdprofit=0.3,
        min_holding_period=20, min_price=1000, min_volume=1000, min_transprice=10000000000,
        min_score_buy=None, max_score_buy=None, min_score_hold=None, max_score_hold=None,
        fee=0.00015, tax=0.0023, slippage=0
    ):
        # settings
        self.start_date = start_date  # 시작일
        self.end_date = end_date  # 종료일
        self.initial_balance = initial_balance  # 초기 자본금
        self.max_investment = max_investment  # 종목 당 최대 투자금
        self.max_holdings = max_holdings  # 최대 보유 종목 개수
        self.max_holdings_value = max_holdings_value  # 보유 종목 최대 가치
        self.rank_threshold_buy = rank_threshold_buy  # 매수 고려할 최저 랭크
        self.rank_threshold_hold = rank_threshold_hold  # 홀드 고려할 최저 랭크
        self.stoploss = stoploss  # 손절 임계값
        self.stopprofit = stopprofit  # 익절 임계값
        self.holdprofit = holdprofit  # 홀드 임계값
        self.min_holding_period = min_holding_period  # 최소 보유 기간
        self.min_price = min_price  # 최소 주가
        self.min_volume = min_volume  # 최소 거래량
        self.min_transprice = min_transprice  # 최소 거래대금
        self.min_score_buy = min_score_buy  # buy를 고려할 최소 점수
        self.max_score_buy = max_score_buy  # buy를 고려할 최대 점수
        self.min_score_hold = min_score_hold  # 보유 유지로 고려할 최소 점수
        self.max_score_hold = max_score_hold  # 보유 유지로 고려할 최대 점수
        self.min_score_buy = 0.28
        self.max_score_buy = 0.7
        self.min_score_hold = 0.25
        self.max_score_hold = 0.5
        self.fee = fee  # 매수/매도 수수료
        self.tax = tax  # 매도 세금
        self.slippage = slippage  # 슬리피지 패널티

        # backtesting context
        self.dt_date = None  # 날짜
        self.date = None  # 날짜 (문자열)
        self.balance = self.initial_balance  # 보유 현금
        self.stocks = {}  # 종목 랭크
        self.holdings = {}  # {<code:str>: {'name': <str>, 'unitprice': <float>, 'amount': <int>, 'profit': <float>, 'date': <str>}}
        self.history = {}  # {<code:str>: [{'date': <str>, 'action': <str>, 'price': <int>, 'amount': <int>}, ...]}
        self.total_profit = None  # 총 손익률
        self.holdings_value = 0  # holdings 평가금
        self.pv = self.balance  # PV

        # output
        os.makedirs('backtest_data', exist_ok=True)
        self.fout = open(os.path.join('backtest_data', 'backtest.txt'), 'w', encoding='utf-8')

    def adjust_score(self):
        self.stocks['adj'] = 0
        self.stocks['score_adj'] = self.stocks['pred']
        for code, history_items in self.history.items():
            if len(history_items) == 0 or code not in self.stocks.index:
                continue
            
            history = history_items[-1]
            row = self.stocks.loc[code, :]
            print(row['pred'])
            s_c = row['pred']
            s_p = history['score']
            d_c = row['date']
            d_p = history['date']
            p_c = row['close']
            p_p = history['price']
            s = (s_c - s_p) / s_p if s_p > 0 else 0
            d = ((convert_str2date(d_c) - convert_str2date(d_p)).days - 60) / 60
            d = min(d, 0) * 0.1
            # 과거 내역 가격 대비 현재 가격 비율
            p = min((p_p - p_c) / p_c, 0) * d
            # 과거 내역보다 점수가 높고 오래됐고 가격도 낮을수록 높은 점수
            adj = s + d + p
            self.stocks.loc[code, 'adj'] = adj
            self.stocks.loc[code, 'score_adj'] = row['pred'] + adj
        self.stocks = self.stocks.sort_values(by='score_adj', ascending=False)
        self.stocks.loc[:, 'rank_adj'] = np.arange(len(self.stocks)) + 1

    def consider_buy(self):
        # 최대 보유 종목 개수 확인
        if len(self.holdings) >= self.max_holdings or self.holdings_value >= self.max_holdings_value:
            return
        
        
        #print(self.stocks[self.stocks['rank_adj'] <= self.rank_threshold_buy])
        for code, stock in self.stocks[self.stocks['rank_adj'] <= self.rank_threshold_buy].iterrows():
            
            # score가 min_score_buy 보다 작으면 무시
            
            if stock['pred'] < self.min_score_buy or stock['pred'] > self.max_score_buy:
                print(stock['pred'])
                print(self.min_score_buy)
                print(1)
                continue

            # 소형주 무시
            if int(stock['close']) < self.min_price:
                print(2)
                continue

            # 거래량 너무 적으면 무시
            if int(stock['volume']) < self.min_volume:
                print(3)
                continue
            
            # 거래대금이 너무 작으면 무시
            if int(stock['close']) * int(stock['volume']) < self.min_transprice:
                print(4)
                continue

            invest_remained = self.max_investment
            if code in self.holdings:
                # 이미 종목을 보유 중
                h = self.holdings[code]
                
                # 추가 보유 여력 확인
                if h['unitprice'] * h['amount'] + int(stock['close']) > self.max_investment:
                    continue

                # 최소 보유 기간 확인
                if (self.dt_date - convert_str2date(h['date'])).days <= self.min_holding_period:
                    continue

                # 추가 매수
                invest_remained = self.max_investment - (h['unitprice'] * h['amount'])

            amount_added = int(invest_remained / int(stock['close']))

            if amount_added > 0:
                if self.buy(code, stock['name'], int(stock['close']), amount_added):
                    self.add_history(code, stock['name'], 'buy', int(stock['close']), amount_added, score=stock['pred'])

    def consider_sell(self):
        # 평가
        holdings = self.holdings.copy()
        for code, h in holdings.items():
            # stock 획득
            if code not in self.stocks.index:
                continue
            stock = self.stocks.loc[code]

            # update profit
            profit = (int(stock['close']) - h['unitprice']) / h['unitprice']
            h['profit'] = profit
            self.holdings[code]['max_price'] = max(self.holdings[code].get('max_price', 0), int(stock['close']))
            self.holdings[code]['max_profit'] = max(self.holdings[code].get('max_profit', 0), profit)

            # 매도 후 바로 다시 매수하는 것을 방지
            if stock['rank_adj'] <= self.rank_threshold_buy:
                continue

            sell_trigger = None
            # 익절
            if self.stopprofit is not None and profit >= self.stopprofit:
                sell_trigger = 'stopprofit'
            # 손절
            elif self.stoploss is not None and profit <= self.stoploss:
                sell_trigger = 'stoploss'
            # 하락 전환 익절
            elif self.stopprofit is not None and self.holdings[code]['max_profit'] - profit >= self.stopprofit / 3 and profit >= 0:
                sell_trigger = 'stopprofit_weak'
            # 청산
            if stock['rank_adj'] > self.rank_threshold_hold and (stock['pred'] < self.min_score_hold or stock['pred'] > self.max_score_hold):
                # 수익권 청산
                if profit >= self.holdprofit:
                    sell_trigger = 'settleprofit'
                # 손실권 청산
                elif profit <= -self.holdprofit:
                    sell_trigger = 'settleloss'
            
            if sell_trigger is not None:
                if self.sell(code, int(stock['close']), h['amount']):
                    self.add_history(code, h['name'], 'sell', int(stock['close']), h['amount'], trigger=sell_trigger, score=stock['pred'], profit=h['profit'])

    def buy(self, code, name, price, amount):
        total = (price * amount) * (1. + self.fee + self.slippage)
        if self.balance < total:
            return False
        h = self.holdings.get(code, {})
        h['name'] = name
        h['unitprice'] = (h.get('unitprice', 0) * h.get('amount', 0) + price * amount) / (h.get('amount', 0) + amount)
        h['amount'] = h.get('amount', 0) + amount
        h['date'] = self.date
        self.holdings[code] = h
        self.balance -= total
        self.update_pv()
        return True

    def sell(self, code, price, amount):
        if code not in self.holdings:
            return False
        h = self.holdings.get(code, {})
        self.holdings[code]['amount'] -= amount
        self.balance += (price * amount) * (1. - (self.tax + self.fee + self.slippage))
        if self.holdings[code]['amount'] <= 0:
            del self.holdings[code]
        else:
            self.holdings[code]['date'] = self.date
        self.update_pv()
        return True

    def update_pv(self):
        self.holdings_value = 0
        for code, h in self.holdings.items():
            if code not in self.stocks.index:
                continue
            stock = self.stocks.loc[code]
            self.holdings_value += int(stock['close']) * h['amount']
        self.pv = self.holdings_value + self.balance

    def add_history(self, code, name, action, price, amount, trigger=None, score=None, profit=None):
        hist = {
            'type': 'history',
            'date': self.date, 
            'pv': self.pv,
            'action': action, 
            'profit': profit,
            'code': code,
            'name': name,
            'price': price, 
            'amount': amount,
            'trigger': trigger,
            'score': score,
            'balance': self.balance,
        }
        self.history[code] = self.history.get(code, []) + [hist]
        self.history[code] = self.history[code][-MAX_HISTORY:]
        print(json.dumps(hist, ensure_ascii=False), file=self.fout, flush=True)

    def get_holding_period(self, code):
        h = self.holdings.get(code)
        if h is None:
            return
        period = (get_today() - convert_str2date(h['date'])).days
        return period

    def report(self, final=False):
        self.update_pv()
        report = {
            'type': 'report' if not final else 'final_report',
            'date': self.date,
            'pv': self.pv,
            'balance': self.balance,
        }
        if final:
            report['holdings'] = self.holdings
        print(json.dumps(report, ensure_ascii=False), file=self.fout, flush=True)

    def run(self):
        print(json.dumps({
            'type': 'setting',
            'start_date': self.start_date,
            'end_date': self.end_date,
            'initial_balance': self.initial_balance,
            'max_investment': self.max_investment,
            'max_holdings': self.max_holdings,
            'max_holdings_value': self.max_holdings_value,
            'rank_threshold_buy': self.rank_threshold_buy,
            'rank_threshold_hold': self.rank_threshold_hold,
            'stoploss': self.stoploss,
            'stopprofit': self.stopprofit,
            'holdprofit': self.holdprofit,
            'min_holding_period': self.min_holding_period,
            'min_price': self.min_price,
            'min_volume': self.min_volume,
            'min_transprice': self.min_transprice,
            'min_score_buy': self.min_score_buy,
            'min_score_hold': self.min_score_hold,
            'fee': self.fee,
            'tax': self.tax,
            'slippage': self.slippage
        }), file=self.fout, flush=True)
        
        #9주 동안 매주 금요일의 데이터를 수집
        for i in tqdm(range((9))):
            self.dt_date = (dt_start_date + timedelta(i)*7)
            self.date = self.dt_date.strftime(FORMAT_DATE)
            path = os.path.join('rankdata',f'df_ranking_{self.date}.csv')

            self.stocks = pd.read_csv(path, dtype={'date': str, 'code': str})
           
            #score의 값에 적당하게 pred값을 조정
            # 원하는 최소값과 최대값을 지정
            desired_min = 0.2
            desired_max = 0.8
            
            #current_min = 0.4
            #current_max = 0.6
            current_min = min(self.stocks['pred'])
            #print('min :'+ str(current_min))
            current_max = max(self.stocks['pred'])
            #print('max :'+ str(current_max))
            #print(self.stocks)
            # Min-Max 스케일링 적용
            self.stocks['pred'] = (self.stocks['pred'] - current_min) / (current_max - current_min) * (desired_max - desired_min) + desired_min
            
            print(self.stocks.sort_values(by='pred', ascending=False)[['code', 'name', 'date', 'pred','close','volume']])

            # score 조정
            self.adjust_score()
            #print(self.stocks)
            # 보유 종목 평가하여 매도 고려
            # 익절, 손절 등 수행
            self.consider_sell()

            # 종목 평가하여 매수 고려
            # 신규 매수, 추가 매수 등 수행
            self.consider_buy()

            # 평가 기록
            self.report()
        
        # 상세 평가 기록
        self.report(final=True)


bt = Backtest(start_date=start_date, end_date=end_date)
bt.run()
```
